7. Utilizando la columna de fecha cread una nueva columna llamada semestre que contenga 1er semestre o 2췈 semestre. Aquellos tickets que pertenezcan a los seis primeros meses tendran el valor de "1er trismestre" y los tickets de julio en adelante tendr치n el valor de "2췈 semestre".

```python
# como no podemos comparar la columna fecha con el tipo dato fecha, vamos a pasarlo a formato datetime

year['fecha'] = pd.to_datetime(year['fecha'])
```

```python
# definimos las condiciones y los valores para utilizar sleect

condiciones = [year['fecha'].dt.month <= 6, year['fecha'].dt.month > 6]

valores = ['1er semestre', '2췈 semestre']

year['semestre'] = np.select(condiciones, valores)
year.head()
```


- Moda

```python
df_wine.mode()
```

- Mediana

```python
df_wine.median()
```

- Media ponderada

```python
w = np.random.randint(0, 100, df_wine.shape[0]) #peso ficticio porque se le asigna el peso con números aleatorios
for col in df_wine.columns:
    avg = np.average(df_wine[col], weights=w)
    print(avg)
```

- Coeficiente de variación

```python
df_wine.std() / df_wine.mean()
```
